```python
def isEven(n):
    if n % 2 == 0:
        return True
        return False
```

```python
def sumIntegers(l):
    total = 0
    for i in l:
        if type(i) == int:
            total += i
    return total
```

```python
sumIntegers([1,2,'Hello', None, 3.5, 4, 3])
```
