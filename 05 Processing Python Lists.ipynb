These are called Lists Comprehensions. Instead of breaking the loop into multiple lines, if the logic is straightforward you should be able to write it down in a single line.

Lambda functions don't typically have any name (other than lambda). Lambda means anonymous or function with no name. 

```python
lambda n: (n * (n + 1)) / 2 
```

In this case we have converted lambda function to named funciton by using this approach. The advantage of creating lambda functions is that turns your code shorter, it's ideal for temporary functions and are more "declarative" and you don't need to write return. Nevertheless, these are not always easy to read and if you take more then 5 seconds to understand them is probably because they're not properly designed (and should have been an user defined function instead - eg. def sumn(n): ). 

```python
sumNl = lambda n: (n * (n + 1)) / 2 
```
