```python
# 1. Create the connection engine
engine = create_engine("postgresql://itversity_sms_user:itversity@localhost:5432/itversity_sms_db")

# 2. Run the query and load directly into a DataFrame
query = "SELECT *" \
"FROM courses " \
"WHERE course_status in ('inactive','draft')"

df_courses = pd.read_sql(query, engine)

# 3. Display the result
df_courses
```

```python
from sqlalchemy import text

query = text("""
    SELECT *
    FROM courses
    WHERE course_name LIKE '%Python%' or course_name LIKE '%SCALA%'
""")

df_courses = pd.read_sql(query, engine)

df_courses
```

```python
from sqlalchemy import text

query = text("""
    SELECT 
             course_author,
             count(*) as course_count
    FROM courses
    WHERE course_status = 'published'
    GROUP BY course_author
""")

df_courses = pd.read_sql(query, engine)

df_courses
```

```python
from sqlalchemy import text

query = text("""
    SELECT 
             course_author,
             count(*) as course_author_count
    FROM courses
    WHERE course_status = 'published'
    GROUP BY course_author
""")

df_courses = pd.read_sql(query, engine)

df_courses
```

```python
from sqlalchemy import text

query = text("""
    SELECT *
    FROM courses
    WHERE course_status = 'draft'
      AND (
            course_name LIKE '%Python%'
         OR course_name LIKE '%Scala%'
      )
""")

df_courses = pd.read_sql(query, engine)
df_courses
```

```python
from sqlalchemy import text

query = text("""
      SELECT 
             course_author,
             count(*) as course_count
      FROM courses
      WHERE course_status = 'published'
      GROUP BY course_author
      HAVING count(*)>1
      
""")

df_courses = pd.read_sql(query, engine)
df_courses
```
