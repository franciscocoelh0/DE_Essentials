```python
def get_column_names(schemas, table_name, sorting_key):
    column_details = schemas[table_name]
    ## return column_details
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    ##return columns
    return [col['column_name'] for col in columns]
```

```python
get_column_names(schemas, 'orders', 'column_position')
```

```python
get_column_names(schemas, 'orders', 'column_name')
```

```python
# If we want it to be sorted interactively

sorting_key = input('Enter the sorting key: ')

def interactive_get_column_names(schemas, table_name, sorting_key = 'column_name'):
    column_details = schemas[table_name]
    ## return column_details
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    ##return columns
    return [col['column_name'] for col in columns]


# I've chosen data_type as the sorting key, but you can choose any key that exists in the column details 
interactive_get_column_names(schemas, 'orders')

# timestamp > string
```
